### *Data Preparation*

```python
df.head()
```

```python
# Drop Unwanted Columns
df = df.drop(columns = ['Ticket ID', 'Customer Name', 'Customer Email','Product Purchased','Date of Purchase','Ticket Subject',
                        'Ticket Description','Resolution','First Response Time', 'Time to Resolution','YearMonth','Age Group'])
```

```python
df.head()
```

```python
# Convert Customer Gender Column as Numeric like Other as 2, Male as 1 and Female as 0
df['Customer Gender'] = df['Customer Gender'].apply({'Other':2,'Male':1, 'Female':0}.get)
```

```python
# Convert Ticket Type Column as Numeric type
df['Ticket Type'] = df['Ticket Type'].apply(
    {'Refund request':5,'Technical issue':4, 'Cancellation request':3,'Product inquiry':2,'Billing inquiry':1}.get)
```

```python
# Convert Ticket Status Column as Numeric type
df['Ticket Status'] = df['Ticket Status'].apply({'Pending Customer Response':2,'Open':1, 'Closed':0}.get)
```

```python
# Convert Ticket Priority Column as Numeric type
df['Ticket Priority'] = df['Ticket Priority'].apply({'Critical':3,'Medium':2,'High':1, 'Low':0}.get)
```

```python
# Convert Ticket Channel Column as Numeric type
df['Ticket Channel'] = df['Ticket Channel'].apply({'Email':4,'Phone':3, 'Social media':2,'Chat':1}.get)
```

```python
df.head()
```

```python
# Check Correlation between them
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

### *Model Building*

```python
X = df.drop(columns = ['Customer Satisfaction Rating'])
y = df['Customer Satisfaction Rating']
```

```python
X
```

```python
y
```

```python
X.values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
X_train
```

```python
y_train
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
